```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # for graphs
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices (2).csv')
```

```python
df
```

```python
plt.scatter(df['area'],df['price'] , color="red", marker="+")
plt.xlabel('area')
plt.ylabel('price')
```

```python
line=LinearRegression()
line.fit(df[['area']], df['price']) #indepencent variable ->[[]]  ,
#dependent variable ->[]
```

```python
#Best m -> slope , coefficient , coef
line.coef_
```

```python
#Best c -> intercept
line.intercept_
```

```python
line.predict([[3800]])# now we can make predicions
```

```python
# another way to solve

#Y_pred = mx+c
#price_pred = m*area+c
135.78767123*3800+180616.43835616432
```

```python
plt.scatter(df['area'],df['price'] , color="red", marker="+")
plt.plot(df['area'],line.predict(df[['area']]))
plt.xlabel('area')
plt.ylabel('price')
```

##Multivarient Linear Regression

```python
df=pd.read_csv('homeprices_multi.csv')
```

```python
df
```

```python
df['bedrooms']=df['bedrooms'].fillna(df['bedrooms'].median())
#mean is prone to outliers, medianis better
```

```python
df
```

```python
line=LinearRegression()
line.fit(df[['area','bedrooms','age']], df['price'])
```

```python
line.coef_
```

```python
line.intercept_
```

```python
line.predict([[2000,3,10]])
```
